```python
import xgboost as xgb
import numpy as np
from sklearn.metrics import accuracy_score
import pandas as pd
from datetime import datetime
from collections import Counter
```

```python
from sklearn.preprocessing import MinMaxScaler
from xgboost import XGBRegressor
from sklearn.metrics import r2_score
import time
```

```python
sample_df=pd.read_csv('encoded_data/sample_data.csv')
```

```python
sample_df.head()
```

```python
X = sample_df.drop(["avg_vote","worlwide_gross_income"], axis=1)
```

```python
X.head()
```

```python
rate_y=sample_df["avg_vote"]
revenue_y=sample_df["worlwide_gross_income"]
```

```python
from sklearn.model_selection import train_test_split
X_rev_train, X_rev_test, revenue_y_train, revenue_y_test = train_test_split(X, revenue_y, test_size=0.2)
X_rat_train, X_rat_test, rate_y_train, rate_y_test = train_test_split(X, rate_y, test_size=0.2)
```

```python
duration_scaler=MinMaxScaler().fit(X[["duration"]])
day_scaler=MinMaxScaler().fit(X[["day"]])
```

```python
X_rev_train_scaled=X_rev_train
X_rev_test_scaled=X_rev_test

X_rat_train_scaled=X_rat_train
X_rat_test_scaled=X_rat_test
```

```python
X_rev_train_scaled["duration"]=duration_scaler.transform(X_rev_train_scaled[["duration"]])
X_rev_train_scaled["day"]=day_scaler.transform(X_rev_train_scaled[["day"]])
X_rev_test_scaled["duration"]=duration_scaler.transform(X_rev_test_scaled[["duration"]])
X_rev_test_scaled["day"]=day_scaler.transform(X_rev_test_scaled[["day"]])

X_rat_train_scaled["duration"]=duration_scaler.transform(X_rat_train_scaled[["duration"]])
X_rat_train_scaled["day"]=day_scaler.transform(X_rat_train_scaled[["day"]])
X_rat_test_scaled["duration"]=duration_scaler.transform(X_rat_test_scaled[["duration"]])
X_rat_test_scaled["day"]=day_scaler.transform(X_rat_test_scaled[["day"]])
```

# Model Selection

```python
from sklearn.linear_model import LinearRegression
start = time.time()
rev_model = LinearRegression()
rat_model = LinearRegression()
rev_model.fit(X_rev_train_scaled, revenue_y_train)
rat_model.fit(X_rat_train_scaled, rate_y_train)
revenue_y_pred = rev_model.predict(X_rev_test_scaled)
rev_score = r2_score(revenue_y_test, revenue_y_pred)
rate_y_pred = rat_model.predict(X_rat_test_scaled)
rate_score = r2_score(rate_y_test, rate_y_pred)
end = time.time()
print(f"LinearRegression model for Revenue Score: {rev_score}")
print(f"LinearRegression model for Rating Score: {rate_score}")
print(f"time spent: {end-start}")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

def create_model():
    model = Sequential()
    model.add(Dense(units=900, activation='relu', input_dim=1870))
    model.add(Dense(units=60, activation='relu'))
    model.add(Dense(units=3, activation='softmax'))
    model.compile(optimizer='adam',
                  loss='mse',
                  metrics=['accuracy'])
    return model

from keras.wrappers.scikit_learn import KerasClassifier
start = time.time()
rev_model = KerasClassifier(build_fn=create_model)
rat_model = KerasClassifier(build_fn=create_model)
rev_model.fit(X_rev_train_scaled, revenue_y_train)
rat_model.fit(X_rat_train_scaled, rate_y_train)
revenue_y_pred = rev_model.predict(X_rev_test_scaled)
rev_score = r2_score(revenue_y_test, revenue_y_pred)
rate_y_pred = rat_model.predict(X_rat_test_scaled)
rate_score = r2_score(rate_y_test, rate_y_pred)
end = time.time()
print(f"DeepLearning model for Revenue Score: {rev_score}")
print(f"DeepLearning model for Rating Score: {rate_score}")
print(f"time spent: {end-start}")
```

```python
from sklearn.ensemble import RandomForestRegressor
start = time.time()
rev_model = RandomForestRegressor()
rat_model = RandomForestRegressor()
rev_model.fit(X_rev_train_scaled, revenue_y_train)
rat_model.fit(X_rat_train_scaled, rate_y_train)
revenue_y_pred = rev_model.predict(X_rev_test_scaled)
rev_score = r2_score(revenue_y_test, revenue_y_pred)
rate_y_pred = rat_model.predict(X_rat_test_scaled)
rate_score = r2_score(rate_y_test, rate_y_pred)
end = time.time()
print(f"RandomForestRegressor model for Revenue Score: {rev_score}")
print(f"RandomForestRegressor model for Rating Score: {rate_score}")
print(f"time spent: {end-start}")
```

```python
from xgboost import XGBRegressor
start = time.time()
rev_model = XGBRegressor()
rat_model = XGBRegressor()
rev_model.fit(X_rev_train_scaled, revenue_y_train)
rat_model.fit(X_rat_train_scaled, rate_y_train)
revenue_y_pred = rev_model.predict(X_rev_test_scaled)
rev_score = r2_score(revenue_y_test, revenue_y_pred)
rate_y_pred = rat_model.predict(X_rat_test_scaled)
rate_score = r2_score(rate_y_test, rate_y_pred)
end = time.time()
print(f"XGBRegressor model for Revenue Score: {rev_score}")
print(f"XGBRegressor model for Rating Score: {rate_score}")
print(f"time spent: {end-start}")
```
